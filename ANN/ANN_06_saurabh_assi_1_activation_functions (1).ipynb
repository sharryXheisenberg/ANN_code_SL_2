# Write a Python program to plot a few activation functions that are being used in neural networks.

# 1.Linear activation function

```python
import numpy as np
import matplotlib.pyplot as plt

def linear(x):
    '''return x '''
    
    return x


x = np.linspace(-10,10)
y = linear(x)

plt.plot(x,y,color = 'red')
plt.title('Linear activation function')
plt.xlabel('Input()')
plt.ylabel('ouyput')
plt.grid(color = 'grey' , linestyle = '--' , linewidth = 0.5)
plt.show()
```

# 2.Binary step activation function

```python
import numpy as np
import matplotlib.pyplot as plt

def binary(x):
    '''Return 0 if x less than 0 else 1'''
    return np.where(x>=0,1,0)


x=np.linspace(-5,5,100)
y = binary(x)

plt.plot(x,y,color = 'green')
plt.title('Binary step activation function')
plt.xlabel('Input(x)')
plt.ylabel('Output')
plt.grid(color = 'grey', linestyle ='--' , linewidth = 0.5)
plt.show()
```

# 3.Sigmoid activation function 

```python
import numpy as np
import matplotlib.pyplot as plt

def Sigmoid(x):
    
    return 1/(1+np.exp(-x))

x=np.linspace(-10,10,100)
y = Sigmoid(x)


plt.plot(x,y,color = 'yellow')
plt.title('Sigmoid activation function')
plt.xlabel('Input(x)')
plt.ylabel('Output')
plt.grid(color = 'grey', linestyle ='--' , linewidth = 0.5)
plt.axhline(0.5 , color = 'red' ,linestyle = '--',label='Threshold(0.5)')
plt.legend()
plt.show()

```

# 6.Softmax activation function 


```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    return np.exp(x)/np.sum(np.exp(x))

X=np.linspace(-5,5,100)
y = softmax(x)

plt.plot(X,y,color = 'indigo')
plt.title('Softmax activation function')
plt.xlabel('Input(x)')
plt.ylabel('Output')
plt.grid(color = 'grey', linestyle ='--' , linewidth = 0.5)
plt.show()
```

# 8.Gelu activation function

```python
import math 
import numpy as np
import scipy



def Gelu(x,a = 0.044715,PI=scipy.pi):
    
    return (x/2)*(1+np.tanh(np.sqrt(2/PI)) * (x+a*x**3))


x=np.linspace(0,1,1000)
y=Gelu(x)

plt.plot(x,y,color = 'red')
plt.title('Gelu activation function')
plt.show()
```

# 11.Leaky relu  

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.01):
    """Leaky ReLU Activation Function"""
    return np.where(x > 0, x, alpha * x)

# Generate data for visualization
x_values = np.linspace(-5, 5, 100)
y_values = leaky_relu(x_values)

# Plot the Leaky ReLU function
plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label='Leaky ReLU(x)', color='orange')
plt.title('Leaky ReLU Activation Function')
plt.xlabel('Input (x)')
plt.ylabel('Output')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

# 12.ELU activation function

```python
import numpy as np
import matplotlib.pyplot as plt

def elu(x, alpha=0.01):
    """Leaky ReLU Activation Function"""
    return np.where(x > 0, x, np.exp(x)-1)

x_values = np.linspace(-5, 5, 100)
y_values = elu(x_values)


plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label='Leaky ReLU(x)', color='orange')
plt.title('ELU Activation Function')
plt.xlabel('Input (x)')
plt.ylabel('Output')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
